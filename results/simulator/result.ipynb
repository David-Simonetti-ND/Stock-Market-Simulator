```python
import matplotlib.pyplot as plt
import re
import numpy as np
```

```python
num_clients = []
pub_time = []
start_int = []
end_int = []
with open("1000clients", "r") as f:
    for line in f:
        line = line.rstrip()
        match = re.search("Num clients: ([0-9]*), Publish Time: ([0-9]*.[0-9]*e-[0-9]*|0\.[0-9]*), Start Int: (0\.[0-9]*), End Int: (0\.[0-9]*)", line)
        num_clients.append(int(match.group(1)))
        pub_time.append(float(match.group(2)))
        start_int.append(float(match.group(3)))
        end_int.append(float(match.group(4)))
    
```

```python
tmp = np.array((num_clients, pub_time, start_int, end_int))
```

```python
np.savetxt("incrementalclients.parsed.csv", tmp.T, delimiter=',')
```

```python
plt.plot(pub_time)
```

```python
plt.plot(start_int)
plt.plot(end_int)
```

```python
num_clients = []
pub_time = []
start_int = []
end_int = []
with open("incrementalclients", "r") as f:
    for line in f:
        line = line.rstrip()
        match = re.search("Num clients: ([0-9]*), Publish Time: ([0-9]*.[0-9]*e-[0-9]*|0\.[0-9]*), Start Int: (0\.[0-9]*), End Int: (0\.[0-9]*)", line)
        num_clients.append(int(match.group(1)))
        pub_time.append(float(match.group(2)))
        start_int.append(float(match.group(3)))
        end_int.append(float(match.group(4)))
    
```

```python

from collections import defaultdict
data = defaultdict(list)
for c, p in list(zip(num_clients,end_int,)):
    if c%100 ==0:
        data[c].append(p)
```

```python
c = []
avg = []
for x, y in data.items():
    c.append(x)
    avg.append(sum(y)/len(y))
```

```python
plt.plot(pub_time)
```

```python
plt.plot(start_int)
plt.plot(end_int)
```

```python
tmp =np.array((c, avg))
```

```python
np.savetxt("incrementalclients.parsed.csv", tmp.T,delimiter=",")
```
